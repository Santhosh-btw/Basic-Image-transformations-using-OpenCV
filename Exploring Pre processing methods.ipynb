```python
import cv2
import matplotlib.pyplot as plt
```

# Raw Image

```python
img = cv2.imread("cloud.jpg")

plt.imshow(img)
```

# Changing color spaces

```python
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_RGB)
```

#  Image resizing 

```python
img_resized = cv2.resize(img, (1920, 1080))

print(f"Original Image Dimensions: {img.shape[1]}x{img.shape[0]}")
print(f"Resized Image Dimensions: {img_resized.shape[1]}x{img_resized.shape[0]}")

cv2.imshow("img_resized", img_resized)
cv2.waitKey(0)
```

# Cropping image 

```python
start_point = [125, 300] # the coordinate to start cropping from (height, width)
end_point = [125, 150] # number of pixels to include in the crop (height, width)

# slicing the needed pixels
img_crop = img[start_point[0]:start_point[0]+end_point[0], start_point[1]:start_point[1]+end_point[1]]
cv2.imshow("img_crop", img_crop)
cv2.waitKey(0)
```

# Grayscaling image 

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(f"Original image dimensions and color channel count: {img.shape}")
print(f"Grayscale image dimensions and color channel count: {img_gray.shape}")

cv2.imshow("img_gray", img_gray)
cv2.waitKey(0)
```
